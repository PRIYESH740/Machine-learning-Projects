```python
from pyECLAT import ECLAT
```

```python
#create an eclat instance 
eclat_instance=ECLAT(data=df,verbose=True)
```

```python
#generate a binary dataframe, that can be used for other analyzes.
eclat_instance.df_bin
```

```python
#a list of all columns of df_bin
eclat_instance.df_bin.columns
```

```python
#a list with all the names of the different items
eclat_instance.uniq_
```

```python
#length of all the names of the different items
len(eclat_instance.uniq_)
```

```python
get_ECLAT_indexes, get_ECLAT_supports = eclat_instance.fit_all(min_support=0.02,
                                                          separator=' & ',
                                                           verbose=True)
```
